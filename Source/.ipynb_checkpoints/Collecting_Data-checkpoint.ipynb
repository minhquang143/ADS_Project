## Introduce Dataset Source

 - Website `fifaindex.com` contains all football players around the world.
 - `fifaindex.com` is a popular website dedicated to providing comprehensive statistical information and data related to the FIFA, specifically the FIFA Ultimate Team (FUT) mode.
 - FIFA Index focuses on delivering detailed player statistics, ratings, and information primarily for FIFA Ultimate Team.
 - The website maintains an extensive database of football players featured in FIFA games, including their in-game attributes, ratings, preferred positions, and other relevant details.
 - Each player has a dedicated profile page containing detailed information such as height, weight, preferred foot, skill moves, weak foot rating, overall ratings and individual attribute ratings.
 - FIFA Index websites also provide market-related information, including player values, prices, and trends within the FIFA Ultimate Team transfer market.

<h2>Start crawling and save collected data</h2>

- The script then iterates through a range of page numbers from 1 to 334 (inclusive). The page number is included in the URL as a parameter.
- In the loop, sends an HTTP GET request to the FIFA Index website for each page using the `requests.get` method.
- The HTML content of the page is parsed using BeautifulSoup.
- Then, finds all the anchor "a" tags with the class "link-player" on the page.
- The script alternately appends the value of the "href" attribute (player profile links) to the get_links list.
- Finally, write the collected player profile links to a text file named "get_links.txt" located in the "../Data/" directory.

```python
url="https://www.fifaindex.com/players"
HEADERS = {"User-Agent": "Mozilla/5.0 (iPad; CPU OS 12_2 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Mobile/15E148"}
get_links=[]
for i in range(1,335):
    count=0
    r=requests.get(url+f"/?page={i}",headers=HEADERS)
    soup=BeautifulSoup(r.text,"html.parser")
    link_players=soup.find_all("a",attrs={"class":"link-player"})
    for link in link_players:
        if link:
            href_value = link.get("href")
        if count==0:  
            get_links.append(href_value)
            count=1
        else:
            count=0
with open("../Data/get_links.txt", "w") as file:
    for item in get_links:
        file.write("%s\n" % item)
```

- For each link got in `get_links.txt`, send a request to `fifaindex.com/ + {link}` to gett HTML element of this link. This HTML about player's information such as Name, Weight, Height, Over ratings, ... all information relevant.
- The HTML content of the page is parsed using BeautifulSoup.
- The `player_info` list includes various player attributes such as preferred foot, birth date, age, and various skill ratings.
- A dictionary named `player_dict` is initialized to organize the collected player information. The keys represent different player attributes, and the corresponding values are initially empty lists.
- `collect_data` has a parameter `course_urls_file`, for each url in courses_url_file, use web scraping to create a dataset of a player profiles.

```python
HEADERS = {"User-Agent": "Mozilla/5.0 (iPad; CPU OS 12_2 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Mobile/15E148"}
player_info = [ 
                "Preferred Foot ",
                "Birth Date ",
                "Age ",
                "Ball Control ",
                "Dribbling ",
                "Marking ",
                "Slide Tackle ",
                "Stand Tackle ",
                "Aggression ",
                "Reactions ",
                "Att. Position ",
                "Interceptions ",
                "Vision ",
                "Composure ",
                "Crossing ",
                "Short Pass ",
                "Long Pass ",
                "Acceleration ",
                "Stamina ",
                "Strength ",
                "Balance ",
                "Sprint Speed ",
                "Agility ",
                "Jumping ",
                "Heading ",
                "Shot Power ",
                "Finishing ",
                "Long Shots ",
                "Curve ",
                "FK Acc. ",
                "Penalties ",
                "Volleys ",
                "GK Positioning ",
                "GK Diving ",
                "GK Handling ",
                "GK Kicking ",
                "GK Reflexes "]
player_dict = { "Name ": [],
                "Height ":[],
                "Weight ":[],
                "Preferred Foot ":[],
                "Birth Date ":[],
                "Age ": [],
                "Nation": [],
                "Club" : [],
                "League" : [],
                "Preferred Positions ":[],
                "OVR " : [],
                "POT " : [],
                "Value " : [],
                "Wage ":[],
                "Ball Control ": [],
                "Dribbling ": [],
                "Marking ": [],
                "Slide Tackle ": [],
                "Stand Tackle ": [],
                "Aggression ": [],
                "Reactions ": [],
                "Att. Position ": [],
                "Interceptions ": [],
                "Vision ": [],
                "Composure ": [],
                "Crossing ": [],
                "Short Pass ": [],
                "Long Pass ": [],
                "Acceleration ": [],
                "Stamina ": [],
                "Strength ": [],
                "Balance ": [],
                "Sprint Speed ": [],
                "Agility ": [],
                "Jumping ": [],
                "Heading ": [],
                "Shot Power ": [],
                "Finishing ": [],
                "Long Shots ": [],
                "Curve ": [],
                "FK Acc. ": [],
                "Penalties ": [],
                "Volleys ": [],
                "GK Positioning ": [],
                "GK Diving ": [],
                "GK Handling ": [],
                "GK Kicking ": [],
                "GK Reflexes ": []
}
def collect_data(course_urls_file):
    url_file = open(course_urls_file)
    urls = url_file.readlines()
    urls_filtered = [item[:-1] for item in urls]
    
    Name = []
    Height = []
    Weight = []
    Nation = []
    Club = []
    League = []
    Preferred_Foot=[]
    Birth_Date=[]
    Age= []
    Preferred_Positions=[]
    OVR = []
    POT = []
    Value = []
    Wage=[]
    Ball_Skills= []
    Defence= []
    Mental= []
    Passing= []
    Physical= []
    Shooting= []
    Goalkeeper= [] 
    for url in urls_filtered:
        a = "https://www.fifaindex.com" + url
        r = requests.get(a,verify=True,headers=HEADERS)
        soup = BeautifulSoup(r.text,"html.parser")
        h1_tag = soup.find("h1")
        
        if h1_tag.span:
            h1_tag.span.decompose()
        
        name = h1_tag.text
        Name.append(name)
        height_weight = soup.find_all("span",attrs = {"class":"data-units-metric"})
        Height.append(height_weight[0].text)
        Weight.append(height_weight[1].text)
        nation = soup.find_all("a",attrs={"class":"link-nation"})
        Nation.append(nation[1].text)
        team_link = soup.find_all("div",attrs={"class":"col-12 col-sm-6 col-lg-6 team"})
        for i in range(len(team_link)):
            if "Joined Club" in team_link[i].text:
                club = team_link[i].find_all("a",attrs={"class":"link-team"})
                Club.append(club[1].text)
                league_link = club[0]['href']
                league = requests.get(f"https://www.fifaindex.com{league_link}",headers=HEADERS)
                league_soup = BeautifulSoup(league.text,"html.parser")
                league_text = league_soup.find("a",attrs={"class":"link-league"})
                League.append(league_text.text)
                break                
        temp = soup.find_all("p", class_="")

        for i in temp:
            temp_text = i.contents[0]
            if temp_text=="Preferred Positions ":
                temp_position=i.find_all("span",attrs={"class":"position"})
                positions_list = [span.text.strip() for span in temp_position]
                positions = ", ".join(positions_list)
                temp_list = player_dict[temp_text]
                temp_list.append(positions)
                player_dict[temp_text] = temp_list
            if temp_text in player_info:
                span_tag = i.find("span")
                span_content = span_tag.text.strip() if span_tag else None
                temp_list = player_dict[temp_text]
                temp_list.append(span_content)
                player_dict[temp_text] = temp_list
                
        check_text = soup.find_all("h5", class_="card-header")[0].text
        if check_text == "FIFA 22 Rankings ":
            h5_tag = soup.find_all("h5", class_="card-header")[1]
        else:
            h5_tag = soup.find_all("h5", class_="card-header")[0]
        rating_spans = h5_tag.find_all("span", class_=re.compile(r"\b(?:badge|badge-dark|rating)\b"))
        ratings = [span.text for span in rating_spans]
        OVR.append(ratings[0])
        POT.append(ratings[1])
        value_tag = soup.find("p", {"class" : "data-currency data-currency-euro"})
        if value_tag:
            content_inside_value = value_tag.contents[1]
            Value.append(content_inside_value.text.replace("€", ""))
            wage_tag = soup.find_all("p", {"class" : "data-currency data-currency-euro"})[1]
            content_inside_wage = wage_tag.contents[1]
            Wage.append(content_inside_wage.text.replace("€", ""))
        else:
            Value.append("None")
            Wage.append("None")
    player_dict["Name "] = Name
    player_dict["Height "] = Height
    player_dict["Weight "] = Weight
    player_dict["Nation"] = Nation
    player_dict["Club"] = Club
    player_dict["League"] = League
    player_dict["OVR "] = OVR
    player_dict["POT "] = POT
    player_dict["Value "] = Value
    player_dict["Wage "] = Wage
    data = pd.DataFrame(player_dict)
    data.rename(columns=lambda x: x.rstrip(), inplace=True)
    return data
```

- Scraping with url in `get_links.txt` created before.

```python
data_footballer = collect_data("../Data/get_links.txt")
display(data_footballer)
```
